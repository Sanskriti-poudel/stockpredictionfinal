```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler


fundamental_folder = "data//fundamental"
fund_files = [file for file in os.listdir(fundamental_folder) if file.endswith("xlsx")]
fund_path = os.path.join(fundamental_folder, fund_files[0])
df1 = pd.read_excel(fund_path, index_col="Symbol")
print(df1.head())
```

```python
df1.index
```

```python
df1.columns
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

```python
numeric_cols = df1.select_dtypes(include=['float64', 'int64']).columns
scaler = MinMaxScaler()
df1[numeric_cols] = scaler.fit_transform(df1[numeric_cols])

print("\nData after normalization (first 5 rows):")
print(df1.head())
```

```python
#EDA
#Correaltion Matrix

corr=df1[numeric_cols].corr()
print(corr)

```

```python
# Histogram of numeric columns
df1[numeric_cols].hist(bins=20, figsize=(12,8))
plt.suptitle('Distribution of Numeric Features')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Fundamental Data')
plt.show()
```

```python
import pandas as pd
import os

data_folder = "data/historical"  
all_stocks = {}

for file in os.listdir(data_folder):
    if file.endswith(".xlsx") and not file.startswith("~$"):
        symbol = os.path.splitext(file)[0]
        df = pd.read_excel(os.path.join(data_folder, file))
        all_stocks[symbol] = df

for symbol, data in all_stocks.items():
    print(f"\n--- {symbol} ---")
    print(data.head(2))  
```

```python
data.index
```

```python
data.columns
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python

# --- EDA for each stock ---

for symbol, df in all_stocks.items():
    print(f"\n--- {symbol} ---")
    print(df.head(2))
    print(df.describe())

    # Plot Close price over time
    if 'Date' in df.columns and 'LTP' in df.columns:
        plt.figure(figsize=(10,4))
        plt.plot(df['Date'], df['LTP'])
        plt.title(f"{symbol} Last Traded Price")
        plt.xlabel("Date")
        plt.ylabel("LTP")
        plt.grid(True)
        plt.show()
    else:
        print(f"Cannot plot Close price for {symbol} due to missing columns.")
        
```
